```python
# Import Dependences
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Retrieve file path
file_path = Path("Recources\cancer patient data sets.xlsx")
```

```python
# Read File
df = pd.read_excel(file_path)
```

```python
# Check file content
df.head()
```

```python
# Replace numerical values with labels
df['Gender'] = df['Gender'].replace({1: 'male', 2: 'female'})

# Display the updated DataFrame
df
```

```python
# Set index to dataframe
df = df.set_index('Patient Id')
```

```python
# PNG File storage prep
folder_path = 'images'
file_name = 'dietary_habits.png'

# Pie chart of Dietary Habits
pie_chart = df['Balanced Diet'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, title='Dietary Habits among Cancer Patients')
plt.show

# Storing PNG
file_path = os.path.join(folder_path, file_name)
plt.savefig(file_path)
```

```python
# PNG File storage prep
folder_path = 'images'
file_name = 'Cancer_Patients_by_Gender.png'

# Pie Chart Patients by gender
pie_chart = df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, title='Cancer Patients by Gender')
plt.show

# Storing PNG
file_path = os.path.join(folder_path, file_name)
plt.savefig(file_path)
```

```python
# PNG File storage prep
folder_path = 'images'
file_name = 'Obesity_rates.png'

# Pie Chart of Obesity rates among cancer Patients
pie_chart = df['Obesity'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, title='Obesity rates among cancer Patients')
plt.show

# Storing PNG
file_path = os.path.join(folder_path, file_name)
plt.savefig(file_path)
```

```python
# Retrive the 'Age' column to prep for binning
for_binning = df['Age']
for_binning
```

```python
# Create new dataframe for binning
binning_df = pd.DataFrame(for_binning)
binning_df
```

```python
# Create bins
bins = [0, 20, 30, 40, 50, 60, 70, 80]

# Create group names
group_names = ["0-20", "21-30", "31-40", "41-50", "51-60", "61-70", "71-80"]

# Create new dataframe with bins
binning_df['Age_range'] = pd.cut(df['Age'], bins, labels=group_names)
binning_df
```

```python
# Create new dataframe categorized by age ranges
age_range_df = pd.DataFrame(binning_df["Age_range"].value_counts())
age_range_df
```

```python
# PNG File storage prep
folder_path = 'images'
file_name = 'bargraph_age_range.png'

# Using the new dataframe to plot age ranges among cancer patients
age_range_df.plot(kind='bar', y='count')
plt.title('Ages Among Cancer Patients')
plt.xlabel('Age Range')
plt.ylabel('Count')

# Storing PNG
file_path = os.path.join(folder_path, file_name)
plt.savefig(file_path)

# Displaying graph
plt.show()
```
